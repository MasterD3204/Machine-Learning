## Subplots

một matplotlib figure có thể bao gồm nhiều subplot

subplot(n_row, n_column, index)
- n_row là số hàng chứa biểu đồ
- n_column là số cột chứa biểu đồ
- index là vị trí của biểu đồ hiện tại

````python
x = np.linspace(-1.4, 1.4, 30)
plt.subplot(2,2,1)
plt.plot(x,x)
plt.subplot(2,2,2)
plt.plot(x, x**2)
plt.subplot(2,2,3)
plt.plot(x, x**3)
plt.subplot(2,2,4)
plt.plot(x,x**4)
plt.tight_layout()
plt.show()
````

có thể dùng subplots() với cú pháp  
fig, axes = plt.subplots(n_rows, n_cols,sharex=False figsize=(width, height))
- fig: đối tượng chứa toàn bộ figure
- axes: mảng các trục (axes) - mỗi trục là 1 biểu đồ nhỏ
- sharex: có chia sẻ trục x giữa nhiều subplot không

````python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
fig, axes = plt.subplots(2, 2, figsize=(10, 6))

# Biểu đồ 1
axes[0, 0].plot(x, y1, color='blue')
axes[0, 0].set_title("Hàm sin(x)")

# Biểu đồ 2
axes[0, 1].plot(x, y2, color='red')
axes[0, 1].set_title("Hàm cos(x)")

# Biểu đồ 3
axes[1, 0].plot(x, np.tan(x), color='green')
axes[1, 0].set_title("Hàm tan(x)")

# Biểu đồ 4
axes[1, 1].plot(x, np.exp(-x), color='purple')
axes[1, 1].set_title("Hàm exp(-x)")

plt.tight_layout()
plt.show()
````

Lựa chọn subplot2grid để đặt biểu đồ một cách linh hoạt hơn  
cú pháp:  
plt.subplot2grid(shape, loc, rowspan=1, colspan=1, fig=None, **kwargs)  
Trong đó:
- shape là kích thước của lưới
- loc là xác định vị trí bắt đầu của subplot trong lưới
- rowspan: số hàng mà subplot chiếm (mặc định là 1)
- colspan: số cột mà subplot chiếm  (mặc định là 1)


````python
plt.subplot2grid((3,3),(0,0),rowspan=2,colspan = 2)
plt.plot(x,  x**2)
plt.subplot2grid((3,3),(0,2))
plt.plot(x,x**3)
plt.subplot2grid((3,3),(1,2), rowspan=2)
plt.plot(x, x**4)
plt.subplot2grid((3,3),(2,0),colspan=2)
plt.plot(x,x**5)
plt.show()
````

## draw text

plt.text(x, y, s, fontdict=None, withdash=False, **kwargs)
- x,y: vị trí đặt văn bản trên biểu đồ (theo toạ độ)
- s : nội dung văn bản ( có thể theo định dạng TeX)
- fontsize
- color
- horizontalalignment / ha: căn chỉnh ngang ('center', 'left', 'right')
- vericalalignment / va : căn chỉnh dọc ('top','bottom', 'center', 'baseline')
- rotation: xoay chữ theo độ (0 đến 360)

````python
x = np.linspace(-1.5, 1.5, 30)
px = 0.8
py = px**2
````

````python
plt.plot(x, x**2, "b-", px, py, "ro")
plt.text(0,1.5, "square function\n$y = x^2$", fontsize= 20, color = 'b',ha = 'center')

plt.text(px - 0.08, py, "Beautiful point", ha ="right", weight = "heavy")
plt.text(px + 0.05, py - 0.4, "x = %0.2f\ny = %0.2f"%(px,py),rotation = -30,color = "gray")
````

thêm chú thích vào đồ thị bằng annotate  
plt.annotate(text, xy, xytext, arrowprops,..)
- text: nội dung cần chú thích
- xy: toạ độ điểm cần chú thích, dạng (x,y)
- xytext: vị trí đoạn văn cần chú thích (mặc định là gần xy)
- arrowprops: thuộc tính của mũi tên nối giữa xy và xytext (facecolor, arrowstyle)

````python
plt.plot(x ,x**2, px,py, "ro")
plt.annotate("Beautiful Point", xy=(px,py), xytext=(px-1.3, py+0.5), color = "green",weight = "heavy", fontsize = 14,
             arrowprops={"facecolor": "lightgreen"} )
````

có thể thêm một hộp bao quanh text bằng cách sử dụng bbox

````python
plt.plot(x, x**2, px, py, "ro")

bbox_props = dict(boxstyle="rarrow,pad=0.3", ec="b", lw=2, fc="lightblue")
plt.text(px-0.2, py, "Beautiful point", bbox=bbox_props, ha="right")

bbox_props = dict(boxstyle="round4,pad=1,rounding_size=0.2", ec="black",
                  fc="#EEEEFF", lw=5)
plt.text(0, 1.5, "Square function\n$y = x^2$", fontsize=20, color='black',
         ha="center", bbox=bbox_props)

plt.show()
````

## Legend

legend dùng để thêm chú giải vào biểu đồ để giải thích ý nghĩa của từng đường, điểm hoặc vùng dữ liệu

phải đặt label của từng đừng trước, sau đó gọi hàm legend

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')

plt.legend()
plt.show()
````

Một số tham só trong legend

loc: tuỳ chỉnh vị chí của legend:
- 'upper right' góc trên phải
- 'upper left' góc trên trái
- 'lower  right' góc dưới phải
- 'best' chọn vị trí tối ưu

frameon: tuỳ chỉn khung chú giải
- True: hiện khung
- False: ẩn khung

fontsize và ncol:
- fontsize: cỡ chữ của chú giải
- ncol: số cột trong chú giải

````python
x = np.linspace(-1.4, 1.4, 50)
plt.plot(x,x**2, "r--", label = "Square function")
plt.plot(x,x**3, "g-", label = "cube function")
plt.legend(loc = "best")
plt.grid(True)
plt.show()
````

## Các loại biểu đồ

### `plot(x, y, **kwargs)`

- **x, y**: Tọa độ của các điểm trên đồ thị
- **color (hoặc c)**: Màu của đường (ví dụ: `'red'`, `'blue'`, `'#FF5733'`)
- **linewidth (hoặc lw)**: Độ dày của đường
- **linestyle (hoặc ls)**: Kiểu đường (ví dụ: `'-'`, `'--'`, `':'`, `'-.'`)
- **marker**: Ký hiệu đánh dấu từng điểm (`'o'`, `'s'`, `'*'`, `'^'`)
- **markersize (hoặc ms)**: Kích thước marker
- **label**: Nhãn cho đường (dùng với `legend()`)

Ví dụ:
```python
plt.plot(x, y, color='green', linestyle='--', marker='o', markersize=8, label='Dữ liệu 1')


````python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, 'b-*', label='y = sin(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.legend()
plt.grid()
plt.show()
````

## 📊 ** `bar()` — Vẽ biểu đồ cột**


### `bar(x, height, **kwargs)`

- **x**: Vị trí của từng cột
- **height**: Chiều cao của mỗi cột
- **width**: Độ rộng của cột (mặc định là `0.8`)
- **color**: Màu sắc của cột
- **edgecolor**: Màu viền của cột
- **linewidth**: Độ dày viền
- **align**: Căn chỉnh cột (`'center'`, `'edge'`)
- **label**: Nhãn cho cột

Ví dụ:
```python
plt.bar(x, height, color='purple', edgecolor='black', width=0.6, label='Doanh thu')


````python
categories = ['A','B','C','D']
values = [3,7,5,9]

plt.bar(categories, values,color="skyblue",edgecolor = 'black')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar plot')
plt.show()
````

## 📉 **`hist()` — Vẽ biểu đồ histogram**


### `hist(x, **kwargs)`

- **x**: Dữ liệu đầu vào
- **bins**: Số lượng khoảng chia
- **color**: Màu sắc của các thanh
- **edgecolor**: Màu viền của thanh
- **alpha**: Độ trong suốt
- **density**: Chuẩn hóa histogram (`True` hoặc `False`)

Ví dụ:
```python
plt.hist(data, bins=20, color='orange', edgecolor='black', alpha=0.8, density=True)


````python
data = np.random.randn(1000)

plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
````

## 🔵 ** `scatter()` — Vẽ biểu đồ phân tán**

### `scatter(x, y, **kwargs)`

- **x, y**: Tọa độ của các điểm
- **s**: Kích thước của từng điểm
- **c**: Màu sắc của từng điểm
- **marker**: Hình dạng của marker
- **alpha**: Độ trong suốt (từ `0` đến `1`)
- **edgecolors**: Màu viền của điểm
- **linewidths**: Độ dày viền

Ví dụ:
```python
plt.scatter(x, y, s=100, c='red', marker='^', alpha=0.7, edgecolors='black', linewidths=1.5)
